```python
import sys
import os
from os.path import join
import glob

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from IPython.display import display, clear_output

sys.path.insert(0, "/home/2649/repos/ABC-SN/code")
import abcsn_config
import abcsn_training
import data_degrading as dg
import data_plotting as dplt
import data_preparation as dp
import preprocessing

from icecream import ic
from importlib import reload

REPO_DIR = "/home/2649/repos/spec_res"
DATA_DIR = os.path.join(REPO_DIR, "data")
BIANCO_DENOISING_DIR = os.path.join(DATA_DIR, "original_resolution_parquet/bianco_denoising")
```

```python
obsolete_sntypes = ["Ia-csm",  "Ia-pec", "Ib-pec", "Ic-pec", "IIL", "IIn", "II-pec"]
sntypes = [
    "Ia-norm", "Ia-91T", "Ia-91bg", "Iax",
    "Ib-norm", "Ibn", "IIb",
     "Ic-norm", "Ic-broad",
     "IIP",
]
```

```python
# Intake a SN Subtype ID from the old system (an ID between 0 and 16 where 0 
#represented Ia-norm and 16 represented II-pec) and output an ID in the new 
#system (an ID between 0 and 9)
sntypes_conversion = {
    0: 0, # Ia-norm
    1: 1, # Ia-91T
    2: 2, # Ia-91T
    4: 3, # Iax
    6: 4, # Ib-norm
    7: 5, # Ibn
    8: 6, # IIb
    10: 7, # Ic-norm
    11: 8, # Ic-broad
    13: 9, # IIP
}
```

```python
file_PP = join(DATA_DIR, "original_resolution_parquet/df_preprocessed.parquet")
df_PP = pd.read_parquet(file_PP)
obsolete_mask = df_PP["SN Subtype"].isin(obsolete_sntypes).values
df_sndata = df_PP.loc[~obsolete_mask].copy(deep=True)

df_sndata["SN Subtype"] = df_sndata["SN Subtype"].cat.remove_unused_categories()
df_sndata["SN Subtype"] = df_sndata["SN Subtype"].cat.reorder_categories(sntypes)
df_sndata["SN Maintype"] = df_sndata["SN Maintype"].cat.reorder_categories(["Ia", "Ib", "Ic", "II"])
df_sndata.index = df_sndata.index.remove_unused_categories()

df_sndata["SN Subtype ID"] = df_sndata["SN Subtype ID"].map(sntypes_conversion)
```
