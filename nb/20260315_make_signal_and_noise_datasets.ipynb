```python
wvl, spectra, df_meta = unpack_dataset(df)

num_spectra = spectra.shape[0]
num_wvl = wvl.shape[0]
df_meta
```

```python
SNRobjs = np.empty(num_spectra, dtype=object)
for i in tqdm(range(num_spectra)):
    metadata_i = df_meta.loc[i].copy(deep=True)
    spectrum_i = spectra[i]

    specsnr = ms.SpectrumSNR(
        metadata_i["SN Name"],
        metadata_i["SN Subtype"],
        metadata_i["Spectrum Phase"],
        wvl,
        spectrum_i)
    specsnr.execute_algorithm(metadata_i)
    SNRobjs[i] = specsnr
```

```python
spectra_signal = np.empty_like(spectra)
spectra_noise = np.empty_like(spectra)
spectra_SNR = np.empty(num_spectra)
spectra_S = np.empty(num_spectra)
spectra_N = np.empty(num_spectra)
for i, specsnr in enumerate(tqdm(SNRobjs)):
    spectra_signal[i] = specsnr.signal
    spectra_noise[i] = specsnr.noise
    spectra_SNR[i] = specsnr.SNR
    spectra_S[i] = specsnr.S
    spectra_N[i] = specsnr.N
```

```python
df_meta.loc[:, "SNR"] = spectra_SNR[...]
df_meta.loc[:, "S (SNR)"] = spectra_S[...]
df_meta.loc[:, "N (SNR)"] = spectra_N[...]
```

```python
df_meta_signal_copy = df_meta.copy(deep=True)
df_meta_noise_copy = df_meta.copy(deep=True)

df_signal_data = pd.DataFrame(data=spectra_signal, columns=wvl.astype(str))
df_noise_data = pd.DataFrame(data=spectra_noise, columns=wvl.astype(str))

df_signal = pd.concat([df_meta_signal_copy, df_signal_data], axis=1)
df_noise = pd.concat([df_meta_noise_copy, df_noise_data], axis=1)
```

```python
df["SNR"] = df_meta["SNR"]
df["S (SNR)"] = df_meta["S (SNR)"]
df["N (SNR)"] = df_meta["N (SNR)"]
```
